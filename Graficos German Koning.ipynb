```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
powerGeneration = pd.read_csv('PowerGeneration.csv')
```

```python
df = powerGeneration.head(50)
```

# Graficos con pyplot

# Gráfico de barras

### Este gráfico es útil para comparar valores entre diferentes columnas. Por ejemplo, puedes utilizarlo para comparar la capacidad total bajo mantenimiento con la capacidad planificada de mantenimiento y la capacidad forzada de mantenimiento

```python
df.plot(kind='bar', x='Power Station', y=['Total Cap. Under Maintenace (MW)', 'Planned Maintanence (MW)', 'Forced Maintanence(MW)'])
plt.title("Comparación de capacidad bajo mantenimiento")
plt.xlabel("Estación de energía")
plt.ylabel("Capacidad (MW)")
plt.show()
```

# Gráfico de línea

### Este gráfico es útil para visualizar cambios en los valores a lo largo del tiempo. Por ejemplo, puedes utilizarlo para ver cómo cambian los valores de "Programa o esperado (MU)" y "Actual (MU)" a lo largo del tiempo.

```python
df.plot(kind='line', x='Dates', y=['Programme or Expected(MU)', 'Actual(MU)'])
plt.title("Cambios en programa vs actual")
plt.xlabel("Fechas")
plt.ylabel("Unidades (MU)")
plt.show()
```

# Gráfico de dispersión

### Este gráfico es útil para visualizar la relación entre dos columnas numéricas. Por ejemplo, puedes utilizarlo para ver la relación entre "Actual (MU)" y "Exceso (+) / Deficiencia (-)".

```python
df.plot(kind='scatter', x='Actual(MU)', y='Excess(+) / Shortfall (-)')
plt.title("Relación entre actual y exceso/deficiencia")
plt.xlabel("Actual (MU)")
plt.ylabel("Exceso (+) / Deficiencia (-)")
plt.show()
```

# Graficos usando seaborn

# Graficos de barra

```python
sns.barplot(x='Power Station', y='Total Cap. Under Maintenace (MW)', hue='Forced Maintanence(MW)', data=df)
plt.title("Comparación de capacidad bajo mantenimiento")
plt.xlabel("Estación de energía")
plt.ylabel("Capacidad (MW)")
plt.show()
```

# grafico de linea

```python
sns.lineplot(x='Dates', y='Programme or Expected(MU)', data=df)
sns.lineplot(x='Dates', y='Actual(MU)', data=df)
plt.title("Cambios en programa vs actual")
plt.xlabel("Fechas")
plt.ylabel("Unidades (MU)")
plt.show()
```

# Grafico de dispersion

```python
sns.scatterplot(x='Actual(MU)', y='Excess(+) / Shortfall (-)', data=df)
plt.title("Relación entre actual y exceso/deficiencia")
plt.xlabel("Actual (MU)")
plt.ylabel("Exceso (+) / Deficiencia (-)")
plt.show()
```
